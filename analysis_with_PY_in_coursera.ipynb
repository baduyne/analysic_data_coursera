### **2.2 Dealing with Missing value**

**How to Deal with Missing Values**

1. **Check with the Data Collection Source**  
   Verify with the original data source to understand if the missing values are intentional or if additional data can be collected.

2. **Drop Missing Values**  
   - **Drop the Variable**: Remove columns that contain a high percentage of missing values if they are not critical to the analysis.
   - **Drop the Data Entry**: Remove rows with missing values, especially when they are few and do not impact the dataset significantly.

3. **Replace Missing Values**  
   - **With a Numerical Value**: Replace missing values with an average.
   - **With Categorical Data**: Replace missing values with the most frequent category (mode).
   - **Based on Other Functions**: Replace missing values using more advanced techniques (e.g., regression imputation, KNN imputation).
   - **Leave as Missing**: Retain missing values if they cannot be imputed accurately, depending on the analysis.

### **2.4 Data Normalization**

**Cách chuẩn hoá dữ liệu**
- Simple Feature scaling:
 $$x_{new} = \frac{x_{old}}{x_{max}}$$
- Min-Max:
 $$ x_{new} = \frac{x_{old} - x_{min}}{x_{max} - x_{min}}$$
- Z_score:
 $$x_{new} = \frac{x_{old} - μ }{σ}$$
    μ: mean
    
    σ(standard deviation): std


- Simple Feature scaling

````python
data = pd.read_csv("gpa.csv")
data.head()
numerical_columns = data.select_dtypes(include=[np.number]).columns
data[numerical_columns] = data[numerical_columns] / data[numerical_columns].max()
print(data)
````

- **Min-Max**

````python
data = pd.read_csv("gpa.csv")
data.head()
numerical_columns = data.select_dtypes(include=[np.number]).columns
data[numerical_columns] = (data[numerical_columns] - data[numerical_columns].min())  / (data[numerical_columns].max()-data[numerical_columns].min())
print(data)
````

- **Z_score**

````python
data = pd.read_csv("gpa.csv")

numerical_columns = data.select_dtypes(include=[np.number]).columns
data[numerical_columns] = (data[numerical_columns] - data[numerical_columns].mean()) / data[numerical_columns].std()
print(data)
````

### **3.2 Correlation - Statistics**

- Pearson Correlation: measure the strength of the correlation between tow features.
    + Correlation coefficient
       
       - Close to +1: Large Positive relationship.
       - Close to -1: Large Negative relationship.
       - Close to 0: No relationship.
    + P-value:

       - P-value < 0.001 ( otherwise know as zero point oh oh one) strong certainty in the result.
       - P-value < 0.05 Moderate( vừa phải ) certainty in the result.
       -  P-value < 0.1 Weak ( /wik/Yếu)  certainty in the result.
       - P-value > 0.1 No certainty in the result.

  Dưới đây là công thức tính hệ số tương quan Pearson và P-value:

- Hệ số tương quan Pearson (r):

Công thức tính hệ số tương quan Pearson giữa hai biến (X) và (Y) là:
$$ r = \frac{n(\sum xy) - (\sum x)(\sum y)}{\sqrt{[n\sum x^2 - (\sum x)^2]}}
$$
P-value:

Để tính P-value cho hệ số tương quan Pearson, bạn có thể sử dụng công thức sau:
$$
t = \frac{r\sqrt{n-2}}{\sqrt{1-r^2}}$$

Sau đó, bạn tra cứu giá trị P-value từ bảng phân phối t với (n-2) bậc tự do.

Nếu bạn cần thêm thông tin hoặc ví dụ cụ thể về cách áp dụng những công thức này trong phân tích dữ liệu, hãy cho tôi biết nhé!

- Syntax:
  ```python
  pear_coef, p_value = stats.personr(df['col'], df['othercol'])
    ```

### **3.2 Correlation - Statistics**

- Chi-square : uesed to determine if there are a significant association bettween two categorical variables.
- Null Hypothesis: kiểm định 2 giá trị này là ngẫu nhiên k có sự liên quan gì đến nhau.
- Alternative Hypothesis: Giả sử rằng có mối liên hệ đáng kể giữa các biến, chỉ ra rằng sự khác biệt quan sát được không chỉ do ngẫu nhiên.
- Calculated:
$$ X^2 = ∑\frac{(O_i - E_i )^2}{E_i}$$
  + $O_i$ giá trị của hàng thứ i.
  + $E_i = \frac{row total * columns}{grand total}$
$$d_f = (r-1)*(c-1)$$
$$ x^2 < chi-square(alpha,d_f) $$
thì k bác bỏ H0 => Không có mối quan hệ giữa 2 biến này

#### **Hand-on: Chi-square**

````python
d = np.array([[20,30],[25,25]])
sum = d.sum()
sum_like_dislike = d[0]+d[1]
sum_male = d[0].sum()
sum_female = d[0].sum()
Ei_male = sum_like_dislike* sum_male/sum
Ei_female = sum_like_dislike* sum_female/sum

Ei = pd.DataFrame({
    "male": Ei_male,
    "female": Ei_female
    
})
Ei_1 = d - Ei
Ei_1 = Ei_1**2
Ei = Ei_1/ Ei
Ei
x = Ei.sum()
x = x.sum()

````

````python
df = (d.shape[0]-1 )*(d.shape[1]-1)
df
````

````python
import scipy.stats as stats

# Đặt giá trị alpha và df
alpha = 0.05  # Mức ý nghĩa (thường là 0.05 cho kiểm định thống kê)


# Tính giá trị chi bình phương tương ứng
chi_square_value = stats.chi2.ppf(1 - alpha, df)

if x < chi_square_value :
    print("fail to reject H0 => không có mỗi quan hệ")
else :
    print(" reject H0 =>  có mỗi quan hệ")
````

Regression (SLR) and Multiple Linear Regression (MLR) is key in data analysis. Here’s a concise comparison:

- Simple Linear Regression (SLR)

  + Definition: SLR involves one independent variable (predictor) and one dependent variable (target).
  + Equation: The relationship is modeled as: $$ y = b_0 + b_1 \cdot x $$
  + Use Case: SLR is used when you want to understand the relationship between two variables. For example, predicting car prices based on highway miles per gallon.
- Multiple Linear Regression (MLR)

  + Definition: MLR involves two or more independent variables and one dependent variable.
  + Equation: The relationship is modeled as: $$ y = b_0 + b_1 \cdot x_1  + b_2 \cdot x_2 + ... + b_n \cdot x_n $$
  + Use Case: MLR is used when multiple factors influence the dependent variable. For example, predicting car prices based on highway miles per gallon, age of the car, and make.

- When to Use Each

  + SLR: Use SLR when you have a single predictor and want to analyze its effect on the target variable.
  + MLR: Use MLR when you have multiple predictors that you believe collectively influence the target variable.

#### **4.1.2 Đánh giá bằng residual plot**

- Định nghĩa: Biểu đồ phần dư là biểu đồ thể hiện giá trị phần dư (residuals) trên trục tung và giá trị dự đoán (fitted values) trên trục hoành.
- Mục đích: Kiểm tra xem các phần dư có phân bố ngẫu nhiên xung quanh đường y = 0 hay không.
- Nhận xét:
    + Phân bố ngẫu nhiên: Nếu phần dư phân bố ngẫu nhiên xung quanh trục y = 0, mô hình có thể là phù hợp.
    + Xu hướng hoặc mô hình: Nếu có xu hướng rõ ràng (ví dụ: đường cong hoặc dải hội tụ/mở rộng), mô hình có thể chưa phù hợp hoặc không tuân theo giả định tuyến tính.
    + Heteroscedasticity (Độ phân tán không đồng nhất): Nếu phần dư tăng dần hoặc giảm dần theo giá trị dự đoán, cần xem xét điều chỉnh (ví dụ: biến đổi dữ liệu).
